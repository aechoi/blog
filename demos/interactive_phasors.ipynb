Visualizations of phasors for sinusoidal waves. See the [tutorial](../blog/phasors.ipynb) for more information.

## Phasors for a single sinusoid
$$
s(t, x) = M \cos(\omega t - k x + \phi)
$$

## Phasors for a sum of two sinusoids (absolute)
$$
s(t, x) = M_1 \cos(\omega_1 t - k_1 x + \phi_1) + M_2 \cos(\omega_2 t - k_2 x + \phi_2)
$$

## Phasors for a sum of two sinusoids (relative)
$$
s(t, x) = M \cos{(\omega t - k x + \phi)} + M |\Gamma| \cos{(\omega  \Omega t - k K x + \phi + \angle\Gamma)}
$$
